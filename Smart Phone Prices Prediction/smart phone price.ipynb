```python
mode_other = df.loc[(df['Processor_Brand'] == 'Other') & 
                    (df['Processor_Series'] != 'Unknown'),
                    'Processor_Series'].mode()[0]  # mode() returns a Series, take first element

# Fill unknown Processor_Series for brand 'Other' with the mode
df.loc[(df['Processor_Brand'] == 'Other') & 
       (df['Processor_Series'] == 'Unknown'),
       'Processor_Series'] = mode_other
```
